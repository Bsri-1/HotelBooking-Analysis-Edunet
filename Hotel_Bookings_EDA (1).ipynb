Now we are going to deal the null values in the company column i.e 112593 containg country codes so using "**MODE**" is preferrable.

Finding the mode initially

```python
hotels.country.mode().to_string()
```

```python
df['country'].fillna(hotels.country.mode().to_string(), inplace=True)
```

Now *Children* Column has the count of children then we replaced missing values with mean values

```python
df['children'].fillna(round(hotels.children.mean()), inplace=True)
```

In the 'agent' column we have also get 16340 so we have replaced it with 0

```python
df['agent'] = df['agent'].fillna(0.0)
```

In the data set some rows have zero guests including adults,children and babies
These rows have no use which can be dropped

```python
df[(df.adults+df.babies+df.children)==0].shape
```

```python
df=df.drop(df[(df.adults+df.babies+df.children)==0].index)
```

Check the types of datatypes and converting it into int64

```python
resort = df[(df['hotel'] == 'Resort Hotel') & (df['is_canceled'] == 0)]
City = df[(df['hotel'] == 'City Hotel') & (df['is_canceled'] == 0)]
```

```python
data_resort = resort[resort['is_canceled'] == 0]
```

```python
data_city = City[City['is_canceled'] == 0]
```

```python
resort_hotel = data_resort.groupby(['arrival_date_month'])['adr'].mean().reset_index()
resort_hotel
```

```python
city_hotel = data_city.groupby(['arrival_date_month'])['adr'].mean().reset_index()
city_hotel
```

```python
final=resort_hotel.merge(city_hotel,on='arrival_date_month')
final.columns=['month','price_for_resort','price_for_city_hotel']
final
```

1)Hotels type ratio::
The database devided into two types of hotels "city" hotels and "Resort" hotels.
The following pie chart visually represents the relative share in percentages of each type of hotel

```python
sns.set_style('darkgrid')
plt.rcParams['font.size'] = 10
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['figure.facecolor']
```

```python
sns.set_style("whitegrid")
hotels_type = df['hotel'].value_counts()
hotels_type
plot = hotels_type.plot.pie(x = 'City Hotel', y='Resort Hotel', autopct='%1.1f%%', figsize=(5, 5))
```

The pie chart reveals that 66% of the hotels in the dataset are city hotels and the other 34% are resort hotels.

2)which month has maximum arrival in the year ?

```python
plt.figure(figsize=(12,4))
sns.countplot(x='arrival_date_month', hue = 'hotel', data= df ,palette= ['r','g'])
plt.title('Months of Arrival')
plt.show()
```

 From the above graph it is observed that Generally, the months of November to March are off-peak periods for hoteliers.
  
 Between June and the end of August, attendance is at its peak, it is the high season. Seasonal downturns are experienced by everyone involved in the journey.
